## 1. Temat pracy doktorskiej
*Predykcja cen akcji polskich przedsiębiorstw przy użyciu wydźwięku komunikatów w języku polskim*

Badanie ma na celu predykcję cen akcji polskich firm. Hybrydowy model zostanie wytrenowany na danych giełdowych oraz wydźwięku wiadomości dotyczących danej firmy zebranych z mediów społecznościowych. Wyniki badania mogą być wykorzystane do estymowania wyników finansowych przy użyciu ogólnodostępnych danych.


###Dlaczego taki temat?
Ceny akcji można przewidzieć (McQueen, G., & Thorley, S., 1991; Bley, J., 2011; Iyke, B. N., & Ho, S. Y., 2021) i im więcej danych jest dostępnych dla modelu, tym lepiej można je oszacować (Pedersen, L. H., 2015; Li, X., Wu, P., & Wang, W., 2020). 

Jako że ceny akcji mogą być traktowane jako zmienna w czasie,  modele stworzone do analizy szeregów czasowych osiągają dobre rezultaty. Najlepsze wyniki realizują modele hybrydowe (Faruk, D. Ö., 2010; Jing, N., Wu, Z., & Wang, H., 2021). 

Modelowanie cen akcji różni się w zależności od rynku. Na ceny akcji wpływa prawo danego kraju, przez co modele mogą uzyskiwać różne wyniki w zależności od kontekstu (Jing, N., Wu, Z., & Wang, H., 2021). Dodatkowo, źródła danych oraz gotowe narzędzia są ograniczone do pewnej puli języków (Yang, H., Luo, L., Chueng, L. P., Ling, D., & Chin, F., 2019). 


###Czy to ma sens?
Powstało sporo prac skupiających się na badaniu wydźwięku z anglojęzycznych mediów, jednak nie skupiają się one na badaniu wyników finansowych konkretnego przedsiębiorstwa i opinii zamieszczanych na jego temat.

## 2. Analiza tekstu a język polski


###Tokenization (tokenizacja)
Jest procesem rozbijania podanego tekstu w przetwarzaniu języka naturalnego na najmniejszą jednostkę w zdaniu zwaną tokenem. Zazwyczaj używa się " " jako znaku oddzielającego słowa i na tym etapie często usuwa się także znaki interpunkcyjne.

Brzmi prosto, ale komplikacje pojawiają się przy tekstach bogatych w skróty lub wyrazy składające się z członów połączonych myślnikiem. Również czasem nie-słowa mogą mieć wpływ na chociażby wydźwięk tekstu (!, emotikony - których powstają osobne [słowniki wydźwięku](http://kt.ijs.si/data/Emoji_sentiment_ranking/)).

Dodatkowo, dla lepszej analizy ten etap może uwzględniać usuwanie słów nie mających znaczenia dla badań - np. spójników. Istnieją przykładowe listy tego typu wyrażeń, natomiast w zależności od rodzaju analizy niektóre ze słów bez nacechowania mogą wnosić inną wartość (np. wspierać analizę logicznych części zdania).


###Stemming (tematyzacja)
Jest procesem obcinania początku lub końca słów w celu dotarcia do formy podstawowej słowa. Działanie jest proste i często wykorzystuje się je dla szybkości, jednak ma sporo wad, które mogą prowadzić do stworzenia słów nie mających związku z oryginalnym tekstem.

###Lemmatization (lematyzacja)


Lematyzacja ma na celu znalezienie podstawowej formy słowa oraz dotarcie do znaczenia danego słowa. Jest to bardziej skomplikowany proces niż stemming.

Celem lematyzacji jest sprowadzenie form fleksyjnych do wspólnej formy bazowej. W przeciwieństwie do tematyzacji, lematyzacja nie polega po prostu na odcinaniu fleksji. Zamiast tego używa leksykalnych baz wiedzy, aby uzyskać prawidłowe podstawowe formy słów.
