## 1. Temat pracy doktorskiej
*Predykcja cen akcji polskich przedsiębiorstw przy użyciu wydźwięku komunikatów w języku polskim*

Badanie ma na celu predykcję cen akcji polskich firm. Hybrydowy model zostanie wytrenowany na danych giełdowych oraz wydźwięku wiadomości dotyczących danej firmy zebranych z mediów społecznościowych. Wyniki badania mogą być wykorzystane do estymowania wyników finansowych przy użyciu ogólnodostępnych danych.


###Dlaczego taki temat?
Ceny akcji można przewidzieć (McQueen, G., & Thorley, S., 1991; Bley, J., 2011; Iyke, B. N., & Ho, S. Y., 2021) i im więcej danych jest dostępnych dla modelu, tym lepiej można je oszacować (Pedersen, L. H., 2015; Li, X., Wu, P., & Wang, W., 2020). 

Jako że ceny akcji mogą być traktowane jako zmienna w czasie,  modele stworzone do analizy szeregów czasowych osiągają dobre rezultaty. Najlepsze wyniki realizują modele hybrydowe (Faruk, D. Ö., 2010; Jing, N., Wu, Z., & Wang, H., 2021). 

Modelowanie cen akcji różni się w zależności od rynku. Na ceny akcji wpływa prawo danego kraju, przez co modele mogą uzyskiwać różne wyniki w zależności od kontekstu (Jing, N., Wu, Z., & Wang, H., 2021). Dodatkowo, źródła danych oraz gotowe narzędzia są ograniczone do pewnej puli języków (Yang, H., Luo, L., Chueng, L. P., Ling, D., & Chin, F., 2019). 


###Czy to ma sens?
Powstało sporo prac skupiających się na badaniu wydźwięku z anglojęzycznych mediów, jednak nie skupiają się one na badaniu wyników finansowych konkretnego przedsiębiorstwa i opinii zamieszczanych na jego temat.

##Analiza tekstu a język polski


###Tokenization (tokenizacja)
Jest procesem rozbijania podanego tekstu w przetwarzaniu języka naturalnego na najmniejszą jednostkę w zdaniu zwaną tokenem. Zazwyczaj używa się " " jako znaku oddzielającego słowa i na tym etapie często usuwa się także znaki interpunkcyjne.

Brzmi prosto, ale komplikacje pojawiają się przy tekstach bogatych w skróty lub wyrazy składające się z członów połączonych myślnikiem. Również czasem nie-słowa mogą mieć wpływ na chociażby wydźwięk tekstu (!, emotikony - których powstają osobne [słowniki wydźwięku](http://kt.ijs.si/data/Emoji_sentiment_ranking/)).

Dodatkowo, dla lepszej analizy ten etap może uwzględniać usuwanie słów nie mających znaczenia dla badań - np. spójników. Istnieją przykładowe listy tego typu wyrażeń, natomiast w zależności od rodzaju analizy niektóre ze słów bez nacechowania mogą wnosić inną wartość (np. wspierać analizę logicznych części zdania).


###Stemming (tematyzacja)
Jest procesem obcinania początku lub końca słów w celu dotarcia do formy podstawowej słowa. Działanie jest proste i często wykorzystuje się je dla szybkości, jednak ma sporo wad, które mogą prowadzić do stworzenia słów nie mających związku z oryginalnym tekstem.

###Lemmatization (lematyzacja)


Lematyzacja ma na celu znalezienie podstawowej formy słowa oraz dotarcie do znaczenia danego słowa. Jest to bardziej skomplikowany proces niż stemming.

Celem lematyzacji jest sprowadzenie form fleksyjnych do wspólnej formy bazowej. W przeciwieństwie do tematyzacji, lematyzacja nie polega po prostu na odcinaniu fleksji. Zamiast tego używa leksykalnych baz wiedzy, aby uzyskać prawidłowe podstawowe formy słów.

```python
search_url = "https://api.twitter.com/2/tweets/search/all"

def get_params(keyword, start_time, end_time, max_results):
  query_params = {'query': keyword,
                      'start_time': start_time,
                      'end_time': end_time,
                      'max_results': max_results,
                    # 'expansions': 'author_id,in_reply_to_user_id,geo.place_id',
                      'tweet.fields': 'id,text,author_id,in_reply_to_user_id,geo,conversation_id,created_at,lang,public_metrics,referenced_tweets,reply_settings,source',
                      'user.fields': 'id,name,username,created_at,description,public_metrics,verified',
                    # 'place.fields': 'full_name,id,country,country_code,geo,name,place_type',
                      'next_token': {}}
  return query_params


def bearer_oauth(r):
    r.headers["Authorization"] = f"Bearer {bearer_token}"
    r.headers["User-Agent"] = "v2FullArchiveSearchPython"
    return r


def connect_to_endpoint(url, params):
    response = requests.request("GET", search_url, auth=bearer_oauth, params=params)
    print(response.status_code)
    if response.status_code != 200:
        raise Exception(response.status_code, response.text)
    return response.json()



def main():
    for key in accountids:
      accountids[key]
      url = search_url
      params = get_params(key, start_time, end_time, max_results)
      json_response = connect_to_endpoint(url, params)
      with open('dataMention.json', 'w', encoding='utf-8') as f:
        json.dump(json_response, f, ensure_ascii=False, indent=4)
      data = json.load(open('/content/dataMention.json'))
      if data["meta"]["result_count"] != 0:
        df = pd.DataFrame(data["data"])
        df.to_csv('SearchTweet{name}{enddate}.csv'.format(name=key, enddate=end_time), encoding='utf-8')
      else:
        print('{name} does not have related Tweets'.format(name=key))


if __name__ == "__main__":
    main()
```

###Analiza tekstu
Korzystam z biblioteki [lpmn-client](https://wiki.clarin-pl.eu/pl/nlpws/query) stworzonej przez CLARIN do obsługi zapytań do poszczególnych narzędzi do analizy tekstu.


Etapy:
1.   [any2txt](https://wiki.clarin-pl.eu/pl/nlpws/services/any2txt) - konwersja pliku zawierającego tekst na tekst
2.   [wsd](https://wiki.clarin-pl.eu/pl/nlpws/services/wsd) - Usługa przeprowadzająca proces ujednoznaczniania znaczeń słów w tekstach polskich i angielskich (sprowadzanie ich do lematów, lematyzacja). Bazuje na zasobach Słowosieci oraz algorytmach przetwarzania sieci semantycznych.
3.   [ccl_emo](https://wiki.clarin-pl.eu/pl/nlpws/services/ccl_emo) - Usługa do analizy wydźwięku i emocji w tekstach. Dla znaczeń pobierana jest informacja emotywna ze Słowosieci-EMO. Dla każdego ze znaczeń pobierana jest informacja o jego polaryzacji wydźwięku oraz fundamentalnych wartościach ludzkich (Puzynina). Nie wszystkie jednostki Słowosieciowe posiadają informację emotywną - w takich sytuacjach jednostka jest pomijana.

ccl_emo zwraca polaryzację wydźwięku per słowo w skali -2, -1, 0, 1, 2. W celu ustalenia wydźwięku dla całego komunikatu, wydźwięk dla poszczególnych słów jest sumowany.

```python
import pandas as pd
import numpy as np

import os

from lpmn_client import download_file, upload_file
from lpmn_client import Task
```

```python
df = pd.read_csv("/content/SearchTweetPaczkomatyPL('2022-01-12T07:05:13.009Z',).csv")
df['sentiment'] = None
df = df.drop(columns=['Unnamed: 0'])
```

```python
for i, row in df.iterrows():
    unique_id = i
    val=(df['text'].values[i])
    with open('plik.txt', 'w') as f:
      f.write(val)
    file_id = upload_file("/content/plik.txt")
    t = Task('any2txt|wcrft2|wsd|ccl_emo({"lang":"polish"})|ccl_emo_stats({"lang":"polish", "split_paragraphs": false}) ')
    output_file_id = t.run(file_id)
    downloaded = download_file(output_file_id, "./out")

    try:
      df_sentiment=pd.read_csv(downloaded, compression='zip', header=0, sep=';')
      df_sentiment = df_sentiment.replace("None", 0)
      df_sentiment = df_sentiment.replace("amb", 0)
      df_sentiment["Polarity"] = pd.to_numeric(df_sentiment["Polarity"])
      column_sum = df_sentiment["Polarity"].sum()
      df['sentiment'].values[i]= column_sum
      print(df['text'].values[i] + ": " + str(column_sum))
    except:
      df['sentiment'].values[i]= None
      print(df['text'].values[i] + ": no sentiment for value " + str(i) )

df.to_csv('Paczkomatysentyment.csv')
```
